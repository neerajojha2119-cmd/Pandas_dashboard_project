### **DATE TIME FUNCTIONS**

```to_datetime will convert the text into date```

````python
df = pd.DataFrame({'date': ['2/10/2024', '2/11/2000'],
                   'value': [2, 3]})

# MM/DD/YYYY by default, first value is considered as month and second value as day

df['date'] = pd.to_datetime(df['date'])
df
````

````python
df.info()
````

````python
# accessing day
df['date'].dt.day
````

````python
# accessing day
df['date'].dt.month
````

````python
# accessing year
df['date'].dt.year
````

````python
# we can specify that consider day as first

df = pd.DataFrame({'date': ['2/10/2024', '2/11/2000'],
                   'value': [2, 3]})

# by default, first value is considered as month and second value as day

df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df
````

````python
# accessing day
df['date'].dt.day
````

````python
# We can specify the format of date

df = pd.DataFrame({'date': ['2024-04-10 20:30:0', 
                            '2024-7-1 19:45:30']})
df['date'] = pd.to_datetime(df['date'], format="%Y-%d-%m %H:%M:%S")
df
````

````python
df['date'].dt.hour
````

````python
df['date'].dt.minute
````

````python
df['date'].dt.second
````

````python
# We can specify the format of date

df = pd.DataFrame({'date': ['2024-04-10 04:30:0 AM', 
                            '2024-7-1 07:45:30 PM']})
df['date'] = pd.to_datetime(df['date'], format="%Y-%d-%m %I:%M:%S %p") # 12 Hour format denoted by %I and %p for AM/PM 
df
````

#### we can observe that 7 is automatically converted to 24 hour format, % I is 12 hour format and %H is 24 hour format, %p is AM/PM

## **Analyzing 3 or more variables with respect to each other**

#### lmplot
 Seaborn has a method called lmplot that may be used to make a scatter plot with a linear regression line (the "lm" stands for linear model). It lets you see the association between two variables as well as the fitted regression line; it's basically a combination of  a scatter plot and a regression plot.

````python
sns.lmplot(customer,x ='Balance',y='Age', hue = 'Exited', fit_reg=True) 
# fit_reg will allow to create regression line
plt.plot()
````

````python
sns.lmplot(customer,x ='NumOfProducts',y='Age', hue = 'Gender',row = 'Exited', fit_reg=True) 
plt.plot() 
````

## Subplots

````python


# Create a figure and an array of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) # here 2,2 is graph matrix, total graphs will be 2X2 = 4

# Using seaborn to create different plots on each subplot
sns.scatterplot(x='Balance', y='Age', data=customer, ax=axes[0, 0])
sns.histplot(data=customer['Age'], ax=axes[0, 1])
sns.boxplot(x='HasCrCard', y='Age', data=customer, ax=axes[1, 0])
sns.barplot(x='HasCrCard', y='Age', data=customer, ax=axes[1, 1])



# Add titles to each subplot
axes[0, 0].set_title('Scatter Plot')
axes[0, 1].set_title('Histogram')
axes[1, 0].set_title('Boxplot')
axes[1, 1].set_title('Barplot')







# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plot
plt.show()
````

## Facetgrids

FacetGrid allows us to create a grid of subplots based on the unique values of one or more categorical variables in your dataset. Each subplot in the grid represents a subset of the data based on the categories of the variables you specify. This is useful for visualizing relationships between variables across different subsets of the data.

````python



# Using FacetGrid
g = sns.FacetGrid(customer, col="Tenure") 

# it will map visualization function (scatterplot) to each category of company size
g.map(sns.scatterplot, 'CreditScore', "Age") 
plt.show()
````
